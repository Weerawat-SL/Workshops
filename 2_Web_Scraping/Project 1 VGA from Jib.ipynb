```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import sys
import requests

chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome("F:\Github\PlayGround\Web Scraping\chromedriver.exe")

products=[] #List to store name of the product
prices=[] #List to store price of the product
```

```python
driver.get('''https://www.jib.co.th/web/product/product_list/2/51''')

content = driver.page_source
soup = BeautifulSoup(content)

for a in soup.findAll('div', attrs={'class':'col-md-3 col-sm-4 col-xs-6 divboxpro'}):
  name=a.find('span', attrs={'class':'promo_name'})
  price=a.find('p', attrs={'class':'price_total'})
  products.append(name.text)
  prices.append(price.text)

df = pd.DataFrame({'Product Name':products,'Price':prices}) 
df
```

```python
df = pd.DataFrame({'Product Name':products,'Price':prices}) 
df.to_csv('VGA_products.csv', index=False, encoding='utf-8')
```
